```python
from fractions import Fraction

def fill_matrix(A):
    with open('matrix.txt', 'r') as f:
        for line in f:
            r = line.split()
            row = []
            for frac in r:
                row.append(Fraction(frac))
            A.append(row)

def print_fraction_matrix(A, n, m):
    for i in range (n):
        for j in range (m):
            if j == m - 1:
                print('|', end = ' ')
            print(A[i][j], end = ' ')
        print()

def jacobi(T, n, m, iter):
    x = []
    vec = input(f'x0')

    for frac in vec.split():
        x.append(Fraction(frac))

    solutions = []
    solutions.append(x)
    print(f'x0: {x}')

    for k in range (iter):

        xi = []
        for i in range (n):

            xi.append(T[i][0])
            sol_index = 1

            for j in range (1, m):

                if j == i + 1:
                    continue

                #print(f'x{k+1}[{i}] += T[{i}][{sol_index}] * x{k}[{j}]')
                xi[i] += T[i][sol_index] * solutions[k][j - 1]
                sol_index += 1
        
        solutions.append(xi)
        print(f'x{k+1}: {xi}')
    

def gauss_seidel(T, n, m, iter):
    x = []
    vec = input(f'x0')

    for frac in vec.split():
        x.append(Fraction(frac))

    solutions = []
    solutions.append(x)
    print(f'x0: {x}')

    for k in range (iter):

        xi = []
        for i in range (n):

            solutions.append([])
            solutions[k + 1].append(T[i][0])
            sol_index = 1

            for j in range (1, m):

                if j == i + 1:
                    continue

                if(j > i + 1):
                    solutions[k + 1][i] += T[i][sol_index] * solutions[k][j - 1]
                else:
                    solutions[k + 1][i] += T[i][sol_index] * solutions[k + 1][j - 1]

                sol_index += 1
        
        print(f'x{k+1}: {solutions[k + 1]}')


def fill_transition_matrix(T, A):
    for i in range (n):
        row = []
        fac_a = Fraction(1, A[i][i])

        row.append(A[i][m-1] * fac_a)

        for j in range (m-1):
            if j != i:
                row.append(-A[i][j] * fac_a)
            
        T.append(row)

A = []
fill_matrix(A)

n = len(A)
m = len(A[0])

T=[]
fill_transition_matrix(T, A)

print('A:')
print_fraction_matrix(A, n, m)

print('T:')
print_fraction_matrix(T, n, m - 1)

jacobi(T, n, m, 3)
print()
gauss_seidel(T, n, m, 3)
```
