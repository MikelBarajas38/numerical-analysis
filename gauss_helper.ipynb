```python
from fractions import Fraction

def fill_matrix(A):
    with open('matrix.txt', 'r') as f:
        for line in f:
            r = line.split()
            row = []
            for frac in r:
                row.append(Fraction(frac))
            A.append(row)

def print_fraction_matrix(A, n, m):
    for i in range (n):
        for j in range (m):
            if j == m - 1:
                print('|', end = ' ')
            print(A[i][j], end = ' ')
        print()

def row_mult(A, i, K):
    for j in range(len(A[0])):
        A[i][j] = A[i][j] * K

def row_sum_factor(A, row1, row2, K):
    for j in range(len(A[0])):
        A[row2][j] = A[row2][j] + A[row1][j] * K

def swap_rows(A, row1, row2):
    temp = A[row1]
    A[row1] = A[row2]
    A[row2] = temp

def copy_matrix(A):
    copy = []
    for row in A:
        copy.append(row.copy())
    return copy

A = []
fill_matrix(A)

n = len(A)
m = len(A[0])

print_fraction_matrix(A, n, m)

A_history = []
op = 'Y'
while op == 'Y':
    #print('Enter the operation you want to perform:')
    #print('1. Row Multiplication')
    #print('2. Row Sum Factor')
    #print('3. Swap Rows')
    #print('restore to previous state')
    #print('4. Exit')
    choice = int(input('operation: '))
    if choice == 1:
        print('\nEnter the row number and the multiplier:')
        row = int(input('row: '))
        K = Fraction(input('K: '))
        AC = copy_matrix(A)
        A_history.append(AC)
        row_mult(A, row-1, K)
        print(f'R{row} -> {K}R{row}')
    elif choice == 2:
        print('\nEnter the row numbers and the factor:')
        row1 = int(input('row src:'))
        row2 = int(input('row dest: '))
        K = Fraction(input('K: '))
        AC = copy_matrix(A)
        A_history.append(AC)
        row_sum_factor(A, row1-1, row2-1, K)
        print(f'R{row2} -> R{row2} + {K}R{row}')
    elif choice == 3:
        print('\nEnter the row numbers to be swapped:')
        row1 = int(input('row1:'))
        row2 = int(input('row2: '))
        AC = copy_matrix(A)
        A_history.append(AC)
        swap_rows(A, row1-1, row2-1)
        print(f'R{row1} <-> R{row2}')
    elif choice == 4:
        if len(A_history) == 0:
            print('No previous state')
        else:
            A = A_history.pop()
            print('\nPrevious state restored')
    else:
        break

    print_fraction_matrix(A, n, m)
    print('Do you want to continue? (Y/N)')
    op = input('cont? :').upper()
```
